```python
'''Point-Biserial correlations for binary variables
Pearson correlation measures the strength and direction of a linear relationship between two continuous variables. We used Pearson correlation for MntTotal, Age and Income. When we try to understand the relationship between a continuous variable MntTotal and binary variables like education or marital status then we should use Point-Biserial Correlation Point-Biserial Correlation is used to measure the strength and direction of the linear relationship between a binary variable and a continuous variable'''
```

```python
from scipy.stats import pointbiserialr
```

```python
for col in cols_education:
    correlation, p_value = pointbiserialr(ifood_df[col], ifood_df['MntTotal'])
    print(col)
    print('Point Biserial Correlation :',correlation)
    print('p-value',p_value)
```

```python
#Feature Engineering
#The data frame contains 5 columns to reflect marital status. We are going to create a new column 'marital' with values: Divorced, Married, Single, Together, Widow. This column will allow us to draw some additional plots.
```

```python
def mar_status(row):
  if row['marital_Divorced']==1:
    return 'Divorced'
  elif row['marital_Married']==1:
    return 'Married'
  elif row['marital_Single']==1:
    return 'Single'
  elif row['marital_Together']==1:
    return 'Together'
  elif row['marital_Widow']==1:
    return 'Widow'
  else:
    return 'Unknown'

ifood_df['Marital_Status'] = ifood_df.apply(mar_status , axis=1)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Marital_Status', y='MntTotal', data=ifood_df)
plt.title('MntTotal by marital status')
plt.xlabel('Marital status')
plt.ylabel('MntTotal')
```

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0

ifood_df['In_relationship'] = ifood_df.apply(get_relationship, axis=1)
ifood_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_for_clustering = ['Income', 'MntTotal', 'In_relationship']
ifood_scaled = ifood_df.copy()
ifood_scaled[cols_for_clustering] = scaler.fit_transform(ifood_df[cols_for_clustering])
ifood_scaled[cols_for_clustering].describe()
```

```python
#Principal Component Analysis (PCA)from sklearn import decomposition
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(ifood_scaled[cols_for_clustering])
ifood_scaled['pc1'] = pca_res[:,0]
ifood_scaled['pc2'] = pca_res[:,1]

```

```python
#Elbow method

X = ifood_scaled[cols_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia = KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,10), inertia_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
#Silhouette score analysis

from sklearn.metrics import silhouette_score
silhouette_list = []
for K in range(2,10):
    model = KMeans(n_clusters = K, random_state=7)
    clusters = model.fit_predict(X)
    s_avg = silhouette_score(X, clusters)
    silhouette_list.append(s_avg)

plt.figure(figsize=[7,5])
plt.plot(range(2,10), silhouette_list, color=(54 / 255, 113 / 255, 130 / 255))
plt.title("Silhouette Score vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
#The highest silhouette score is for 4 clusters.
```

```python
model = KMeans(n_clusters=4, random_state = 7)
model.fit(ifood_scaled[cols_for_clustering])
ifood_scaled['Cluster'] = model.predict(ifood_scaled[cols_for_clustering])
```

```python
#Visualisation of clusters

plt.figure(figsize=(8, 6))
sns.scatterplot(x='pc1', y='pc2', data=ifood_scaled, hue='Cluster', palette='viridis')
plt.title('Clustered Data Visualization')
plt.xlabel('Principal Component 1 (pc1)')
plt.ylabel('Principal Component 2 (pc2)')
plt.legend(title='Clusters')
```

```python
ifood_df['Cluster'] = ifood_scaled.Cluster
ifood_df.groupby('Cluster')[cols_for_clustering].mean()
```

```python
mnt_data = ifood_df.groupby('Cluster')[cols_mnt].mean().reset_index()
mnt_data.head()
```

```python
melted_data = pd.melt(mnt_data, id_vars="Cluster", var_name="Product", value_name="Consumption")
plt.figure(figsize=(12, 6))
sns.barplot(x="Cluster", y="Consumption", hue="Product", data=melted_data, ci=None, palette="viridis")
plt.title("Product Consumption by Cluster")
plt.xlabel("Cluster")
plt.ylabel("Product Consumption")
plt.xticks(rotation=0)
plt.legend(title="Product", loc="upper right")

plt.show()
```
